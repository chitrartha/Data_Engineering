Web scraping using BS

```python
import requests
import lxml
import pandas as pd
from bs4 import BeautifulSoup
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area').text
soup = BeautifulSoup(website_url, 'lxml')
```

```python
html_page = soup.find("table", class_="wikitable sortable")
#table is tag and class is the attribute. Using class_ because class is keywork in python.
#We ignore the query-tablesorter in teh class
#find the 1st occurence of <table> 'tag' with 'class' attribute as "wikitable sortable"
```

```python
print(html_page.prettify)
```

```python
print (html_page.a)
#Gets the 1st occurence of <a> tag
```

```python
print (html_page.a.text)
#<a> is the tag and text is the content of that tag.
```

```python
html_page.findAll('a')
#findAll finds all the occurence of <a> tag
```

```python
countries_name = []
for i in html_page.findAll('a'):
    countries_name.append(i['title'])
    #title is the 'attribute' of <a> class
print (countries_name)
```

```python
countries_name = []
for i in html_page.findAll('a'):
    countries_name.append(i.text)
    #text is the 'content' of <a> class
print (countries_name)
```

```python
p = html_page.findAll('tr')
print (p)
```

```python
header = list(p[0].text.split("\n"))
#Trying to get the header
print(header)
```

```python
header = ([i for i in header if i != '']) + [None] + [None]
#Removing the blank fields and teh length of header should be equal to max length of any rows.
print (header)
```

```python
#Now, we get the rows from table in to the list of lists.
list_of_lists = []
for i in range(1,len(p)):
    list_of_lists.append(p[i].text.split("\n"))
print (list_of_lists)
```

```python
#As every alternate field will have blank value beacuse of our split logic we remove them.
for c,e  in enumerate(list_of_lists):
    list_of_lists[c].pop(0)
    list_of_lists[c].pop(1)
    list_of_lists[c].pop(2)
    list_of_lists[c].pop(3)
    list_of_lists[c].pop(4)
print (list_of_lists[3]) #Displaying only 1 row.
```
